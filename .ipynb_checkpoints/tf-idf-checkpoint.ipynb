## tf-idf

```python
import math
from collections import defaultdict
corpus = [
    "中国 电力 与华为 签署 战略 合作协议",
    "肇庆市 市场监督管理局 公布 2021民生领域案件 查办“铁拳”行动 典型案例",
    "恒大地产 震动 许家印 卸任 股价重挫",
    "排名第一 平安租赁 入选 “消费者最满意的汽车融资租赁公司” 榜单",
    "汽车之家 宣布 集生态战略 升级",
    "华为周桃园 绿色能源 目标网 助力 运营商 实现 绿色低碳 网络",
    "为高层次 和 紧缺人才 开辟绿色通道 江苏省 349人 被直接认定 高级职称",
    "腾讯 造芯官方 回应来了 科技巨头 入局 概念股 业绩大涨",
    "北交所张梅 构建 长周期 全链条服务 中小企业 市场体系",
    "中小券商 发力 布局 新三板业务",
    "恒大 百亿 旧改 国家队 驰援",
    "部分 供应商 某大商票 造成影响 集中反馈",
    "中航产融公司 所属 子公司 中航",
    "信托 资金 未曾 恒大地产 及 关联企业 发生 贷款 投资的业务",
    "QQ 远去 微信 成为腾讯 元宇宙社交 最大弱点",
    "苹果 谷歌 五巨头 十年 并购 616家 初创企业 小型收购 开始 被反垄断 盯上",
    "中远海运 顺利完成 埃及斋月十日 城铁路 项目 首批 轨道车辆交付",
    "盛业资本 与 中铁资本 保理 深化战略合作 构建 基建行业 供应链 数字金融生态",
    "深度 CDL 1美元 清仓 协信远创 多家国资入局",
    " 雪域先锋号 首次 实现 远程操控 始发",
    "媒体 调查 15家 互联网平台 涉金融产品 少数 存在违规 部分业务 监管标准 待明晰"
]
words = []
for sentence in corpus:
    words.append(sentence.strip().split())
def Counter(words):
    word_count = []
    for sentence in words:
        word_dict = defaultdict(int)
        for word in sentence:
            word_dict[word] += 1
        word_count.append(word_dict)
    return word_count
word_count = Counter(words)
def tf(word, word_dict):
    return word_dict[word] / sum(word_dict.values())
def count_sentence(word, word_count):
    return sum([1 for i in word_count if i.get(word)]) 
def idf(word, word_count):
    return math.log(len(word_count) / (count_sentence(word, word_count) + 1))

def tfidf(word, word_dict, word_count):
    return tf(word, word_dict) * idf(word, word_count)
p = 1
for word_dict in word_count:
    for word, cnt in word_dict.items():
        print("\"{}\":{},".format(word, tfidf(word, word_dict, word_count)))
```
